```python
def frame_size(data, sr, frame_time, frame_shift):
    frameWidth = int(round(frame_time * sr)) #ширина кадра
    frameShift = int(frame_shift * frameWidth) #сдвиг кадра
    df = frameWidth - frameShift # приращение
    frameCount = int(round(data.size / df)) - 1 #количество кадров
    return frameWidth, frameShift, frameCount, df
```

```python
# разбиение речевого сигнала на кадры
def signal_to_cards(y, cad_num, frame_shift, nframes):
    cads = []
    pos = 0
    cad_len = nframes//cad_num #целочисленное деление
    first_cads_num = nframes - cad_len*cad_num
    for i in range(cad_num):
        if i < first_cads_num:
            cads.append(y[pos: pos + int((cad_len + 1) * (frame_shift+1))])
            pos += cad_len + 1
        elif i == cad_num - 1:
            cads.append(y[pos: pos + cad_len])
        else:
            cads.append(y[pos: pos+ int(cad_len * (frame_shift+1))])
            pos += cad_len
    return cads
```

```python
# разбиение речевого сигнала на кадры с настраиваемой степенью перекрытия
time_len = nframes/sr
print('длительноть файла:',time_len, 'сек')
print('длительность кадра:', frame_time*1000, 'м сек')
print('количество кадров:', frameCount)
print('степень перекрытия:', frame_shift)

cads = signal_to_cards(y, frameCount, frame_shift, nframes)
```

```python
def dataToEnergy(data, sr, frame_time, frame_shift):
    frameWidth, frameShift, frameCount, df = frame_size(data, sr, frame_time, frame_shift)
    data = data.astype(float)
    E = np.array(
        [np.sum(data[df * i: df * i + frameWidth] ** 2) / frameWidth for i in range(frameCount)])
    return np.array(E)
```

```python
def alt_vad(cad_energy, window=20):
    z = cad_energy
    trashold=np.mean(z)
    activity = []
    for i in range(len(z)):

        if i < window:
            if np.mean(z[i:i+window]) > trashold:
                activity.append(1)
            else:
                activity.append(0)

        elif window <= i <= len(z)-window:
            if np.mean(z[i-window:i+window]) > trashold:
                activity.append(1)
            else:
                activity.append(0)

        elif i > len(z)-window:
            if np.mean(z[i-window:i]) > trashold:
                activity.append(1)
            else:
                activity.append(0)

    return np.array(activity)
```

```python
# сегментация
def segmentation(data, sr, frame_time, frame_shift, voice_active, Zpor=30):
    Z_ = np.where(data >= 0, 1, 0)

    frameWidth, frameShift, frameCount, df = frame_size(data, sr, frame_time, frame_shift)
    # energy = dataToEnergy(data, sr, frame_time, frame_shift)
    Z = []
    for i in range(frameCount):
        Zi = Z_[i * df: i * df + frameWidth].astype(float)
        Z2 = Zi[1:]
        Z1 = Zi[:-1]
        Z.append(np.sum(np.abs(Z2 - Z1)) / 2)
    Z = np.array(Z)[voice_active]
    indexes = []
    for i in range(1, Z.shape[0] - 1):
        if Z[i] < Zpor < Z[i + 1] or Z[i] > Zpor > Z[i + 1]:
            indexes.append(i)
    return indexes
```

```python

# Фильтрация методом вычитания спектров
def filtering(y, sr, frame_time, frame_shift):
    frameWidth, frameShift, frameCount, df = frame_size(y, sr, frame_time, frame_shift)
    y_without = np.zeros_like(y)
    noise = y[: frameWidth]
    alf = 1.35
    beta = 0.2

    for i in range(int(round(y.size / frameWidth)) - 1):
        yi = y[i*frameWidth: (i + 1)*frameWidth]
        spectr_sig = fft(yi)
        arg = np.angle(spectr_sig)
        spectr_noise = fft(noise[: yi.size])

        delt = np.abs(spectr_sig) - alf * np.abs(spectr_noise)
        tet = beta * np.abs(spectr_noise)
        spectr_res = np.where(delt > 0, delt, tet)
        y_without[i * frameWidth: i * frameWidth + yi.size] = ifft(spectr_res * np.exp(1j * arg))
    return y_without
```

```python

# Фильтр низких частот
def low_pass_filter(y, sr, wsr):
    b, a = sig.butter(4, wsr, 'low', output='ba', fs=sr, analog=False)
    return sig.filtfilt(b, a, y)
```

```python
# Фильтрация
y_n_mvc = filtering(y, sr, frame_time, frame_shift)
sf.write(f'{path_to_folder}\\cleaned_SP.wav', y_n_mvc, sr)

y_n_filter = low_pass_filter(y, sr, 2500)
sf.write(f'{path_to_folder}\\cleaned_filter.wav', y_n_filter, sr)
```
